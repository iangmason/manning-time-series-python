```python
import pandas as pd
import datetime as dt
```

```python
weather = pd.read_csv("hr_temp_20170201-20200131_subset.csv")
weather.head()
```

```python
load = pd.read_csv("hrl_load_metered - 20170201-20200131.csv")
load.head()
```

```python
joined = load.join(weather)
joined.head()
```

```python
joined_dropped_columns = joined[["DATE", "mw", "HourlyDryBulbTemperature"]]
joined_dropped_columns.head()
```

```python
joined_dropped_columns.info()
```

```python
 joined_dropped_columns["DATE"] = pd.to_datetime(joined_dropped_columns["DATE"])
```

```python
joined_dropped_columns["hour"] = joined_dropped_columns["DATE"].dt.hour
```

```python
joined_dropped_columns["weekday"] = joined_dropped_columns["DATE"].dt.weekday
```

```python
joined_dropped_columns["month"] = joined_dropped_columns["DATE"].dt.month
```

```python
joined_dropped_columns["year"] = joined_dropped_columns["DATE"].dt.year
```

```python
joined_dropped_columns
```

```python
mwTemp = joined_dropped_columns.set_index("DATE")
```

```python
mwTemp
```

```python
mwTemp.info()
```

```python
mwTemp["HourlyDryBulbTemperature"] = mwTemp["HourlyDryBulbTemperature"].interpolate(method='linear')
```
